# ***dataFrames***

**A pandas DataFrame is a primary, two-dimensional, table-like data structure used for data manipulation and analysis in Python.**

Key components

A DataFrame is composed of three main parts:

• Data: The actual values contained within the table.

• Rows: Labeled observations (also called the index).

• Columns: Labeled variables that can hold different data types (e.g., numbers, text,
dates).

```python
# Row and Columns
# Creating dataframes with help of dictionary
student_details = {
    "Name" : ['Rahul','Aman','Naman'],
    'Subject': ['Maths','English','social science'],
    'Marks': [50,60,70,],
    'Address': ['Jaipur','Delhi','Madhya Pradesh']
}
pd.DataFrame(student_details)
```

```python
student_detail = [
    ['Rahul','Maths',50,'Jaipur'],
    ['Aman','English',60,'Delhi'],
    ['Naman','social science',70,'Madhya Pradesh']
]
pd.DataFrame(student_detail, columns=['Name','Subject','Marks','Address'])
```
